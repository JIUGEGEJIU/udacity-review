```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import xgboost as xgb
import operator
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt 

from data_preprocess import data_preprocess
from extract_features import extract_features
from rmspe import *
```

```python
train = pd.read_csv("source data/train.csv", parse_dates=[2])
test = pd.read_csv("source data/test.csv", parse_dates=[3])
store = pd.read_csv("source data/store.csv")
```

```python
df_preprocessed = data_preprocess(train, test, store)
```

```python
features = []
df_with_all_features = extract_features(features, df_preprocessed)
```

```python
print(features)
```

```python
df_train = df_with_all_features[df_with_all_features['DataSetFlag'] == 1]
df_test = df_with_all_features[df_with_all_features['DataSetFlag'] == 0]
```

```python
# 选取与test数据集相同时间跨度的数据作为validation data set
TestTimeDuration = df_test.Date.max() - df_test.Date.min()
ValidateMaxDate = df_train.Date.max()
ValidateMinDate = ValidateMaxDate - TestTimeDuration

valid_indices = df_train['Date'].apply(lambda x: (x >= ValidateMinDate and x <= ValidateMaxDate))

train_indices = valid_indices.apply(lambda x: (not x))
```

```python
X_train = df_train[train_indices]
X_valid = df_train[valid_indices]
y_train = np.log1p(df_train['Sales'][train_indices])
y_valid = np.log1p(df_train['Sales'][valid_indices])
```

```python
num_boost_round = 5000
model_num = 0
model_result_dict = {}
```

```python
for max_depth in range(10,15):
    for subsample in range(7,9):
        for colsample_bytree in range(5,7):

            params = {'objective': 'reg:linear',
                       'booster': 'gbtree',
                       'eta': 0.03,
                       'silent': 1,
                       'seed': 909,
                      }


            params['colsample_bytree'] = colsample_bytree * 0.1
            params['subsample'] = subsample * 0.1
            params['max_depth'] = max_depth

            print("train xgboost model-{}".format(model_num))
            dmatrix_train = xgb.DMatrix(X_train[features], y_train)
            dmatrix_valid = xgb.DMatrix(X_valid[features], y_valid)

            watch_list = [(dmatrix_train, 'train'), (dmatrix_valid, 'eval')]
            gbm = xgb.train(params, dmatrix_train, num_boost_round, evals=watch_list, early_stopping_rounds=100, feval=rmspe_xg, verbose_eval=True)

            print('performing validation')
            y_hat = gbm.predict(xgb.DMatrix(X_valid[features]))
            error = rmspe(X_valid.Sales.values, np.expm1(y_hat))
            print('Validation RMSPE: {:.5f}'.format(error))

            params['valid_error'] = error
            model_result_dict[str(model_num)] = params

            dmatrix_test = xgb.DMatrix(df_test[features])
            y_test = gbm.predict(dmatrix_test)

            # output
            result = pd.DataFrame({'Id': test['Id'], 'Sales': np.expm1(y_test)})
            result.to_csv('output/xgb_op_{}.csv'.format(model_num), index=False)

            model_num += 1
```

```python
df_model_result = pd.DataFrame([model_result_dict])
df_model_result = df_model_result.T
```

```python
df_model_result.to_csv('output/xgb_op_dict.csv', index=False)
```
